```python
from mplsoccer.pitch import Pitch, VerticalPitch
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
from highlight_text import HighlightText, ax_text, fig_text
import numpy as np
from matplotlib.cbook import get_sample_data
```

```python
data = pd.read_excel("Firmino Shot Data 2022-23.xlsx")
data['x'] = data['x']*1.2
data['y'] = data['y']*0.8

datasot = data[data['Result'] == 'SoT']
datag = data[data['Result'] == 'Goal']
datam = data[data['Result'] == 'Missed']
```

```python
fig ,ax = plt.subplots(figsize=(15.5,12))
fig.set_facecolor('#181818')
ax.patch.set_facecolor('#181818')

pitch = VerticalPitch(pitch_type='statsbomb', orientation='vertical',
              pitch_color='#181818', line_color='#FFFFFF', figsize=(15, 12),
              constrained_layout=False, tight_layout=True , half=True , label=False)

#pitch = VerticalPitch(half=True)
pitch.draw(ax=ax)

#Plotting Shots on Target
pitch.scatter(datasot.x, datasot.y, s=datasot['xG']*1300,
             marker='o', edgecolors='none', c='#FF5959', zorder=2, ax=ax , alpha=0.45 , ls='-')

pitch.scatter(datasot.x, datasot.y, s=datasot['xG']*1300,
             marker='o', edgecolors='#FF5959', c='none', zorder=3, ax=ax , alpha=1 , ls='-' , lw=2.2)


#Plotting Goals
pitch.scatter(datag.x, datag.y, s=datag['xG']*1300 , marker='o' , edgecolors='#8ff00f',
              c='none' , ax=ax, ls='-' , lw=2.5 , alpha=1 , zorder=4)

pitch.scatter(datag.x, datag.y, s=datag['xG']*1300 , marker='o' , edgecolors='none',
              c='#8ff00f' , ax=ax, ls='-' , lw=2.5 , alpha=0.45 , zorder=4)

#Plotting Missed/Blocked Shots
pitch.scatter(datam.x, datam.y, s=datam['xG']*1300 , marker='o' , edgecolors='#40e0d0',
              c='none' , ax=ax, ls='-' , lw=2.5 , alpha=1 , zorder=4)

pitch.scatter(datam.x, datam.y, s=datam['xG']*1300 , marker='o' , edgecolors='none',
              c='#40e0d0' , ax=ax, ls='-' , lw=2.5 , alpha=0.45 , zorder=4)




pitch.scatter(64*1.2 , 10*0.8 , s=1800 , marker='o' , edgecolors='#FF5959', c='none' , ax=ax, ls='-' ,
              lw=2.5 , alpha=0.9 , zorder=4)
pitch.scatter(64*1.2 , 10*0.8 , s=1800 , marker='o' , edgecolors='none', c='#FF5959' , ax=ax, ls='-' ,
              lw=2.5 , alpha=0.3 , zorder=4)
pitch.annotate('23' , xy=(64*1.2,10*0.8) , xytext=(77.5,6.6) , ax=ax , font='Franklin Gothic Medium' ,
               fontsize=22 , color='#FFFFFF' , fontweight='bold' , zorder=5)
pitch.annotate('Shots' , xy=(64*1.2,10*0.8) , xytext=(77.5,11.2) , ax=ax , font='Franklin Gothic Medium' ,
               fontsize=22 , color='#FFFFFF' , fontweight='bold' , zorder=5)



pitch.scatter(64*1.2 , 30*0.8 , s=1800 , marker='o' , edgecolors='#8ff00f', c='none' , ax=ax, ls='-' ,
              lw=2.5 , alpha=0.9 , zorder=4)
pitch.scatter(64*1.2 , 30*0.8 , s=1800 , marker='o' , edgecolors='none', c='#8ff00f' , ax=ax, ls='-' ,
              lw=2.5 , alpha=0.3 , zorder=4)

pitch.annotate('8' , xy=(64*1.2,30*0.8) , xytext=(77.5,23.2) , ax=ax , font='Franklin Gothic Medium' ,
               fontsize=22 , color='#FFFFFF' , fontweight='bold' , zorder=5)
pitch.annotate('Goals' , xy=(64*1.2,30*0.8) , xytext=(77.5,27.2) , ax=ax , font='Franklin Gothic Medium' ,
               fontsize=22 , color='#FFFFFF' , fontweight='bold' , zorder=5)



pitch.scatter(64*1.2 , 50*0.8 , s=1800 , marker='o' , edgecolors='#FFFFFF', c='none' , ax=ax, ls='-' ,
              lw=2.5 , alpha=0.9 , zorder=4)
pitch.scatter(64*1.2 , 50*0.8 , s=1800 , marker='o' , edgecolors='none', c='#FFFFFF' , ax=ax, ls='-' ,
              lw=2.5 , alpha=0.2 , zorder=4)
pitch.annotate('0.15' , xy=(64*1.2,50*0.8) , xytext=(77.5,38) , ax=ax , font='Franklin Gothic Medium' ,
               fontsize=18 , color='#FFFFFF' , fontweight='bold' , zorder=5)
pitch.annotate('xG per Shot' , xy=(64*1.2,50*0.8) , xytext=(77.5,43.2) , ax=ax , font='Franklin Gothic Medium' ,
               fontsize=22 , color='#FFFFFF' , fontweight='bold' , zorder=5)


pitch.scatter(64*1.2 , 78*0.8 , s=1800 , marker='o' , edgecolors='#FFFFFF', c='none' , ax=ax, ls='-' ,
              lw=2.5 , alpha=0.9 , zorder=4)
pitch.scatter(64*1.2 , 78*0.8 , s=1800 , marker='o' , edgecolors='none', c='#FFFFFF' , ax=ax, ls='-' ,
              lw=2.5 , alpha=0.2 , zorder=4)
pitch.annotate('0.47' , xy=(64*1.2,78*0.8) , xytext=(77.5,60.4) , ax=ax , font='Franklin Gothic Medium' ,
               fontsize=18 , color='#FFFFFF' , fontweight='bold' , zorder=5)
pitch.annotate('xG per 90' , xy=(64*1.2,78*0.8) , xytext=(77.5,65.4) , ax=ax , font='Franklin Gothic Medium' ,
               fontsize=22 , color='#FFFFFF' , fontweight='bold' , zorder=5)


pitch.annotate('*Smaller the dot size, Lower the xG of that Shot' , xy=(88*1.2,28*0.8) , xytext=(84,24.5) , ax=ax , font='Franklin Gothic Medium' ,
               fontsize=14 , color='#FFFFFF' , fontweight='bold' , zorder=5)



pitch.scatter(68*1.2 , 10*0.8 , s=200 , marker='o' , edgecolors='#40e0d0', c='none' , ax=ax, ls='-' ,
              lw=2.5 , alpha=0.9 , zorder=4)
pitch.scatter(68*1.2 , 10*0.8 , s=200 , marker='o' , edgecolors='none', c='#40e0d0' , ax=ax, ls='-' ,
              lw=2.5 , alpha=0.3 , zorder=4)

pitch.annotate('Missed Shots' , xy=(68*1.2,10*0.8) , xytext=(82.1,9.8) , ax=ax , font='Franklin Gothic Medium' ,
               fontsize=14 , color='#FFFFFF' , fontweight='bold' , zorder=5)



pitch.lines(85 , 0 , 85 , 80 , ls='-' , lw=1.5 , color='#FFFFFF' , ax=ax , zorder=1 , alpha=0.8)

plt.gca().invert_yaxis()

fig_text(
    0.512, 0.965, "<Roberto Firmino>", font='Franklin Gothic Medium' ,size=30,
    ha="center", color="#FFFFFF" , fontweight='bold' , highlight_textprops=[{"color": '#FFFFFF'}]
)

fig_text(
    0.512, 0.918,
    "Liverpool | 648 Minutes Played | Shot Map Card | Made by: @pranav_m28",
    font='Franklin Gothic Medium', size=24,
    ha="center", color="#FFFFFF" , fontweight='bold'
)

im = plt.imread('Firmino.png')
newax = fig.add_axes([0.465, 0.75, 0.098, 0.95], anchor='SE', zorder=1)
newax.imshow(im)
newax.axis('off')


#im = plt.imread('lfctr.png')
#newax = fig.add_axes([0.185, 0.77, 0.100, 0.95], anchor='SE', zorder=1 , alpha=0.6)
#newax.imshow(im)
#newax.axis('off')

#plt.savefig("Firmino Shot Map Card.png" , dpi=300, bbox_inches='tight')
```
